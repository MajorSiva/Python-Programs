```python
#1
class Person:
    def __init__(self, name, designation):
        self.name = name
        self.designation = designation

    def learn(self):
        print(self.name, "is learning.")

    def walk(self):
        print(self.name, "is walking.")

    def eat(self):
        print(self.name, "is eating.")


class Programmer(Person):
    def __init__(self, name, designation, company_name):
        super().__init__(name, designation)
        self.company_name = company_name

    def coding(self):
        print(self.name, "is coding at", self.company_name)


class Dancer(Person):
    def __init__(self, name, designation, group_name):
        super().__init__(name, designation)
        self.group_name = group_name

    def dancing(self):
        print(self.name, "is dancing with group", self.group_name)


class Singer(Person):
    def __init__(self, name, designation, band_name):
        super().__init__(name, designation)
        self.band_name = band_name

    def singing(self):
        print(self.name, "is singing with band", self.band_name)

    def play_guitar(self):
        print(self.name, "is playing guitar with band", self.band_name)


if __name__ == "__main__":
    programmer = Programmer("Alice", "Software Engineer", "TechCorp")
    programmer.learn()
    programmer.walk()
    programmer.eat()
    programmer.coding()

    print()

    dancer = Dancer("Bob", "Dance Instructor", "Fusion Crew")
    dancer.learn()
    dancer.walk()
    dancer.eat()
    dancer.dancing()

    print()

    singer = Singer("Charlie", "Lead Vocalist", "Rockstars")
    singer.learn()
    singer.walk()
    singer.eat()
    singer.singing()
    singer.play_guitar()
```

```python
#7
import datetime
df1= pd.read_csv('student_marks.csv')
#df1['DOB']=df1['DOB'].astype('datetime64[ns]')
#a
def calculate_age(dob):
    dob = datetime.datetime.strptime(dob, '%d-%m-%Y')
    today = datetime.datetime.today()
    return today.year - dob.year - ((today.month, today.day) < (dob.month, dob.day))

df1['age'] = df1['DOB'].apply(calculate_age)
print(df1[['Name', 'DOB', 'age']])

#b
def calculate_average(row):
    subjects = ['Maths', 'Physics', 'Chemistry', 'English', 'Biology', 'Economics', 'History', 'Civics']
    marks = [row[subject] for subject in subjects if pd.notna(row[subject])]
    if marks:
        return sum(marks) / len(marks)
    else:
        return 0

def assign_grade(average):
    if average >= 90:
        return 'A'
    elif average >= 80:
        return 'B'
    elif average >= 70:
        return 'C'
    elif average >= 60:
        return 'D'
    else:
        return 'F'

df1['Average'] = df1.apply(calculate_average, axis=1)
df1['Grade'] = df1['Average'].apply(assign_grade)
print(df1)

#c
passed_students= df1[df1['Grade']<='D']
passed_students.to_csv('passed_students.csv',index=False)
print("\nFile Saved Successfully")
```
